```python
import os
import glob
import time
import pandas as pd
import numpy as np

directory = "Raw_Data\Direct Loans"
files = glob.glob(os.path.join(directory, "*.xls"))
files = sorted(files, key=lambda x: os.path.getmtime(x))
year = 1999
year2 = 2000

for filename in files:
    new_name = f"AY{year:04d}-{year2:04d}Direct_Loans.xls"
    
    old_path = filename
    new_path = os.path.join(directory, new_name)
    
    if os.path.exists(new_path):
        print(f"Target file '{new_name}' already exists. Skipping renaming.")
    else:
        os.rename(old_path, new_path)
        print(f"Renamed {filename} to {new_name}")
    
    year += 1
    year2 += 1
```

```python
years2 = []
years = []
dataframes = []

for z in range(2000,2006):
    years2.append(str(z))

for i in range(1999,2005):
    years.append(str(i))

for w,x in zip(years , years2):

    filename = f'Raw_Data/direct_loans/AY{w}-{x}Direct_Loans.xls'

    try:

        df = pd.ExcelFile(filename)
        df1= pd.read_excel(df,'Award Year Summary',skiprows=5)
        df1['Year'] = w
        dataframes.append(df1)
        print(f"{filename} has appended successfuly")
        
    except:
        
        print(f"{filename} did not append")

base = pd.concat(dataframes)
```

```python
years2 = []
years = []
dataframes2 = []

for z in range(2006,2011):
    years2.append(str(z))

for i in range(2005,2010):
    years.append(str(i))

for w,x in zip(years , years2):

    filename = f'Raw_Data/direct_loans/AY{w}-{x}Direct_Loans.xls'

    try:

        df = pd.ExcelFile(filename)
        df2= pd.read_excel(df,'Award Year Summary',skiprows=5)
        df2['Year'] = w
        dataframes2.append(df2)
        print(f"{filename} has appended successfuly")
        
    except:
        
        print(f"{filename} did not append")

base2 = pd.concat(dataframes2)
```

```python
years2 = []
years = []
dataframes3 = []

for z in range(2011,2018):
    years2.append(str(z))

for i in range(2010,2017):
    years.append(str(i))

for w,x in zip(years , years2):

    filename = f'Raw_Data/direct_loans/AY{w}-{x}Direct_Loans.xls'

    try:

        df3 = pd.ExcelFile(filename)
        df3= pd.read_excel(df3,'Award Year Summary',skiprows=5)
        df3['Year'] = w
        dataframes3.append(df3)
        print(f"{filename} has appended successfuly")
        
    except:
        
        print(f"{filename} did not append")

base3 = pd.concat(dataframes3)
```

```python
years2 = []
years = []
dataframes4 = []

for z in range(2018,2023):
    years2.append(str(z))

for i in range(2017,2022):
    years.append(str(i))

for w,x in zip(years , years2):

    filename = f'Raw_Data/direct_loans/AY{w}-{x}Direct_Loans.xls'

    try:

        df4 = pd.ExcelFile(filename)
        df4= pd.read_excel(df4,'Award Year Summary',skiprows=6)
        df4['Year'] = w
        dataframes4.append(df4)
        print(f"{filename} has appended successfuly")

    except:

        print(f"{filename} did not append")

base4 = pd.concat(dataframes4)
```

```python
frames = [base,base2,base3,base4]
direct_loan_df = pd.concat(frames)

direct_loan_df = direct_loan_df.replace('-', np.nan)
direct_loan_df = direct_loan_df.replace(' -', np.nan)
direct_loan_df['Zip Code'] = direct_loan_df['Zip Code'].str.replace(r'\D', '', regex=True)
direct_loan_df['Zip Code'] = direct_loan_df['Zip Code'].str.slice(0,5)

direct_loan_df.to_csv('Raw_Data/direct_loan_df_draft.csv')
```
